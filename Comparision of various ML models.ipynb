```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

## Handling Null Values

```python
df.isnull().sum()
```

```python
df['Age'].unique()
```

```python
df['Age'].mean()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Cabin'].unique()
```

```python
df['Cabin'].mode()
```

```python
df['Cabin'] = df.groupby('Survived')['Cabin'].transform(lambda x: x.fillna(x.mode()[0]))
```

```python
df['Embarked'].unique()
```

```python
df['Embarked'].mode()
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.describe()
```

## Replacing Categorical values

```python
df.replace({ 'Sex' : {'male':0 , 'female':1}},inplace=True)
```

```python
df.replace({'Embarked' : {'S':0,'C':1,'Q':2}}, inplace=True)
```

```python
df
```

## Removing the unnecessary columns 

```python
X = df.drop( columns = ['PassengerId' , 'Name' , 'Ticket','Cabin','Survived'] , axis=1)
Y = df['Survived']
```

```python
print(X)
```

```python
print(Y)
```

## Splitting Dataset

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1 = LogisticRegression(max_iter=1000)
model1.fit(X_train,Y_train)
```

```python
model1_pred=model1.predict(X_test)
print(model1_pred)
```

```python
# score=accuracy_score(model1_pred,Y_test)
print("Accuracy of Logistic Regression: ",accuracy_score(model1_pred,Y_test))
```

```python
confusion_matrix = confusion_matrix(Y_test, model1_pred)
print(confusion_matrix)
print(classification_report(model1_pred,Y_test))
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model2 = DecisionTreeClassifier(random_state=42)
model2.fit(X_train, Y_train)
```

```python
model2_pred = model2.predict(X_test)
```

```python
print("Decision Tree Accuracy:", accuracy_score(Y_test, model2_pred))
print(confusion_matrix(Y_test, model2_pred))
print(classification_report(Y_test, model2_pred))
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model3 = RandomForestClassifier(n_estimators=100, random_state=42)
model3.fit(X_train, Y_train)
```

```python
model3_pred = model3.predict(X_test)
```

```python
print("Random Forest Accuracy:", accuracy_score(Y_test,model3_pred ))
print(confusion_matrix(Y_test,model3_pred))
print(classification_report(Y_test, model3_pred))
```

## K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model4 = KNeighborsClassifier(n_neighbors=5)  # You can tune n_neighbors
model4.fit(X_train, Y_train)
```

```python
model4_pred = model4.predict(X_test)
```

```python
print("K-Nearest Neighbors Accuracy:", accuracy_score(Y_test, model4_pred))
print(confusion_matrix(Y_test,model4_pred))
print(classification_report(Y_test, model4_pred))
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model5 = GaussianNB()
model5.fit(X_train, Y_train)  
```

```python
model5_pred = model5.predict(X_test)
```

```python
print("Naive Bayes Accuracy:", accuracy_score(Y_test, model5_pred))
print(confusion_matrix(Y_test, model5_pred))
print(classification_report(Y_test, model5_pred))
```

## Support Vector Machine (SVM)

## GridsearchCV for Hyperparameter tunning

```python
from sklearn.model_selection import GridSearchCV
```

## For Logistic Regression

```python
param_logreg = {
    'C': [0.1, 1, 10],
    'solver': ['liblinear', 'lbfgs']
}

grid_logreg = GridSearchCV(LogisticRegression(max_iter=1000), param_logreg, cv=5)
grid_logreg.fit(X_train, Y_train)

print("Best Logistic Regression:", grid_logreg.best_params_)
print("Accuracy:", accuracy_score(Y_test, grid_logreg.predict(X_test)))
```

## Decision Tree

```python
param_dt = {
    'max_depth': [3, 5, 10],
    'min_samples_split': [2, 5, 10]
}

grid_dt = GridSearchCV(DecisionTreeClassifier(), param_dt, cv=5)
grid_dt.fit(X_train, Y_train)

print("Best Decision Tree:", grid_dt.best_params_)
print("Accuracy:", accuracy_score(Y_test, grid_dt.predict(X_test)))
```

## Random Forest

```python
param_rf = {
    'n_estimators': [50, 100],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5]
}

grid_rf = GridSearchCV(RandomForestClassifier(), param_rf, cv=5)
grid_rf.fit(X_train, Y_train)

print("Best Random Forest:", grid_rf.best_params_)
print("Accuracy:", accuracy_score(Y_test, grid_rf.predict(X_test)))
```

## Naive Bayes

```python
grid_nb = GridSearchCV(GaussianNB(), param_grid={}, cv=5)
grid_nb.fit(X_train, Y_train)

print("Naive Bayes Accuracy:", accuracy_score(Y_test, grid_nb.predict(X_test)))
```

## Support Vector Machine (SVM)

```python
param_svm = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

grid_svm = GridSearchCV(SVC(), param_svm, cv=5)
grid_svm.fit(X_train, Y_train)

print("Best SVM:", grid_svm.best_params_)
print("Accuracy:", accuracy_score(Y_test, grid_svm.predict(X_test)))
```

## K-Nearest Neighbors (KNN)

```python
param_grid_knn = {
    'n_neighbors': list(range(1, 21)),
    'weights': ['uniform', 'distance']
}

# Initialize the model
knn = KNeighborsClassifier()

# GridSearchCV
grid_knn = GridSearchCV(knn, param_grid_knn, cv=5)
grid_knn.fit(X_train, Y_train)

# Results
print("Best parameters for KNN:", grid_knn.best_params_)
print("Best score for KNN:", grid_knn.best_score_)
```

```python
print("Accuracy of Logistic Regression:", accuracy_score(Y_test, grid_logreg.predict(X_test)))
print("Accuracy of Decision Tree:", accuracy_score(Y_test, grid_dt.predict(X_test)))
print("Accuracy of Random Forest:", accuracy_score(Y_test, grid_rf.predict(X_test)))
print("Accuracy of Naive Bayes:", accuracy_score(Y_test, grid_nb.predict(X_test)))
print("Accuracy of SVM :", accuracy_score(Y_test, grid_svm.predict(X_test)))
print("Best score for KNN:", grid_knn.best_score_)
```
